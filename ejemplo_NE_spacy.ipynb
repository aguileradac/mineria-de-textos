```python
import spacy
```

```python
nlp = spacy.load("es_core_news_sm")
```

```python
doc = "Muchos años después, frente al pelotón de fusilamiento, el coronel Aureliano Buendía había de recordar aquella tarde remota en que su padre lo llevó a conocer el hielo. Macondo era entonces una aldea de veinte casas de barro ycañabrava construidas a la orilla de un río de aguas diáfanas que se precipitaban por un lecho de piedras pulidas, blancas y enormes como huevos prehistóricos. El mundo era tan reciente, que muchas cosas carecían de nombre, y para mencionarlas había que señalarlas con el dedo. Todos los años, por el mes de marzo, una familia de gitanos desarrapados plantaba su carpa cerca de la aldea, y con un grande alboroto de pitos y timbales daban a conocer los nuevos inventos. Primero llevaron el imán. Un gitano corpulento de barba montaraz y manos de gorrión, que se presentó con el nombre de Melquíades, hizo una truculenta demostración pública de lo que él mismo llamaba la octava maravilla de los sabios alquimistas de Macedonia."
doc_anotado = nlp(doc)
```

```python
#imprime las NER encontradas
for ent in doc_anotado.ents:
    print(ent.text, ent.label)
```

```python
#imprime información de los tokens relativa a NER
for token in doc_anotado:
    print(token.text + "\t" + token.pos_ + "\t" + token.ent_iob_+"-"+token.ent_type_)
```
