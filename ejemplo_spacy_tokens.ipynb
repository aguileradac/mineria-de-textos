```python
import spacy
import pandas as pd
```

```python
nlp = spacy.load("es_core_news_sm")
```

```python
doc = nlp(u"Muchos años después, frente al pelotón de fusilamiento, el coronel Aureliano Buendía había de recordar aquella tarde remota en que su padre lo llevó a conocer el hielo.")
```

```python
resultado = pd.DataFrame(columns=['token', 'lema', 'pos', 'tag', 'dep', 'is_stop'])
for token in doc:
    resultado = resultado.append({
        'token': token.text,
        'lema': token.lemma_,
        'pos': token.pos_,
        'tag': token.tag_,
        'dep': token.dep_,
        'is_stop': token.is_stop},
        ignore_index=True)
    
resultado
```
